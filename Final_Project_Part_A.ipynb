**Part A: Before data normalization**





Import of data

```python
import pandas as pd
concrete_data = pd.read_csv('https://cocl.us/concrete_data')
Features_columns = concrete_data.columns
Features = concrete_data[Features_columns[Features_columns != 'Strength']]
Target = concrete_data['Strength']
```

Input data

```python
NExecutions = 50
```

Train-Test split

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(Features, Target, test_size=0.30, random_state=100)
```

Construction of neural network structure

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
NFeatures = X_train.shape[1]
InputShape = (NFeatures,)

# Defining a function to construct the regression model
def regression_model():
    # model construction
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=InputShape))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

from sklearn.metrics import mean_squared_error
MSE = []
for iteration in range (NExecutions):
  # Initiating the model and training
  Model = regression_model()
  Model.fit(X_train,y_train,validation_split=0.3,epochs=50,verbose=2)

  y_predicted = Model.predict(X_test)
  MSE.append(mean_squared_error(y_test, y_predicted))
```

```python
import numpy as np
MSE = np.array(MSE)
Mean_of_MSE = MSE.mean()
STD_of_MSE = MSE.std()
print('Results before data normalization:')
print('Mean of MSEs is {}'.format(Mean_of_MSE))
print('Standard deviation of MSEs is {}'.format(STD_of_MSE))
```

**Part B: After data normalization**

```python
from sklearn import preprocessing
scaler = preprocessing.StandardScaler().fit(X_train)
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
MSE = []
for iteration in range (NExecutions):
  # Initiating the model and training
  Model = regression_model()
  Model.fit(X_train,y_train,validation_split=0.3,epochs=50,verbose=2)

  y_predicted = Model.predict(X_test)
  MSE.append(mean_squared_error(y_test, y_predicted))
```

```python
import numpy as np
MSE = np.array(MSE)
Mean_of_MSE = MSE.mean()
STD_of_MSE = MSE.std()
print('The results obtained after normalization')
print('Mean of MSEs is {}'.format(Mean_of_MSE))
print('Standard deviation of MSEs is {}'.format(STD_of_MSE))
```

**Part C: Effect of increase of number of epochs**

```python
MSE = []
for iteration in range (NExecutions):
  # Initiating the model and training
  Model = regression_model()
  Model.fit(X_train,y_train,validation_split=0.3,epochs=100,verbose=2)

  y_predicted = Model.predict(X_test)
  MSE.append(mean_squared_error(y_test, y_predicted))
```

```python
import numpy as np
MSE = np.array(MSE)
Mean_of_MSE = MSE.mean()
STD_of_MSE = MSE.std()
print('The results obtained by 100 epochs')
print('Mean of MSEs is {}'.format(Mean_of_MSE))
print('Standard deviation of MSEs is {}'.format(STD_of_MSE))
```

**Part D: Effect of increase of number of layers**

```python
# Defining a function to construct the regression model
def regression_model2():
    # model construction
    model = Sequential()
    model.add(Dense(10, activation='relu', input_shape=InputShape))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(10, activation='relu'))
    model.add(Dense(1))
    
    # compile model
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model
```

```python
MSE = []
for iteration in range (NExecutions):
  # Initiating the model and training
  Model = regression_model2()
  Model.fit(X_train,y_train,validation_split=0.3,epochs=50,verbose=2)

  y_predicted = Model.predict(X_test)
  MSE.append(mean_squared_error(y_test, y_predicted))
```

```python
import numpy as np
MSE = np.array(MSE)
Mean_of_MSE = MSE.mean()
STD_of_MSE = MSE.std()
print('The results obtained after increaseing the numbre of layers to 3')
print('Mean of MSEs is {}'.format(Mean_of_MSE))
print('Standard deviation of MSEs is {}'.format(STD_of_MSE))
```
